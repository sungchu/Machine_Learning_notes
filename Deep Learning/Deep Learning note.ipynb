#### 2.2.2 softmax 函數
- $ y_k=\frac{exp(a_k)} {\sum_{i=1}^{n} exp(a_i)} $
   - $ y_k $ 是計算第 k 個輸出
   - $ a_k $ 是輸入的第 k 個訊號
- 輸出的是 0 到 1.0 之間的實數，且輸出的總和為 1，使得 softmax 函數的輸出可以解釋為機率
- 一般而言，神經網路的類別會把相當於輸出最大神經元的類別當作辨識結果。即使套用 softmax 函數，輸出最大神經元的位置不變。神經網路進行分類時，一般會省略輸出層的 softmax 函數。

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

由於指數的值可能會變很大，而造成溢位，因此有了下方的改良方法

$ y_k=\frac{exp(a_k)} {\sum_{i=1}^{n} exp(a_i)} $ 
= $ y_k=\frac{C exp(a_k)} {C\sum_{i=1}^{n} exp(a_i)} $ 
= $ y_k=\frac{C exp(a_k+logC)} {C\sum_{i=1}^{n} exp(a_i+logC)} $
= $ y_k=\frac{C exp(a_k+C')} {C\sum_{i=1}^{n} exp(a_i+C')} $

```python
def softmax(a):
    c = np.max(a) # 為了防範溢位，一般會使用輸入訊號中，最大的數值
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
# 測試資料
## 輸入層的 784(28*28) 個神經元，輸出層的 10 個神經元

import sys, os
sys.path.append(os.pardir) 
import numpy as np
import pickle
from mnist import load_mnist
from functions import sigmoid, softmax


def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=False)
    return x_test, t_test

# 在 init_network()中，載入儲存 pickle 檔案 sample_weight.pkl 中，已經學習完畢的權重參數。
# 在這個檔案中，權重與偏權值的參數會儲存成字典型態的變數。
def init_network():
    with open("sample_weight.pkl", 'rb') as f: # .pkl是 pickle 檔案，此功能可以把值形程式的物件，當作檔案儲存起來
        network = pickle.load(f)
    return network


def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y
```

```python
x, t = get_data()
network = init_network()

accuracy_cnt = 0

for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y) # 取得陣列中最大值的索引值
    if p == t[i]: # 神經網路預測的答案與正確答案標籤比對
        accuracy_cnt += 1

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))
```

```python
# 把上個 cell 的內容改成批次處理的寫法，批次處理可以縮短影像處理的時間
x, t = get_data()
network = init_network()

batch_size = 100 # 批次處理的量
accuracy_cnt = 0

for i in range(0, len(x), batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch,axis = 1) 
    accuracy_cnt += np.sum(p == t[i:i+batch_size])

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))
```

## 3 神經網路的學習
- 神經網路也就是利用資料，自動決定權重參數的值
- 感知收斂定理: 藉由次數有限的學習，解決線性分離問題。但非線性分離問題無法學習
- 機器學習的方法(依序為人為、半人為、無人為)
   1. 資料 > 人類想出的演算法 > 答案
   2. 資料 > 人類想出來的特徵量(SIFT、SURF、HOG等) > 機器學習(SVM、KNN等) > 答案
      - 特徵量是指可以從輸入資料(影像)中，精準擷取出本質資料(重要資料)的轉換器
      - 需要根據問題使用適當的特徵量，不然也無法獲得良好的結果
   3. 資料 > 神經網路(深度學習) > 答案

## 4 損失函數
- 在神經網路的學習中，會根據某個單一指標，表現目前的狀態，並以該指標為基準找尋**最適合的權重參數**，而該指標就稱為損失函數
- 損失函數可以使用任意函數，但主要使用均方誤差或交叉熵誤差
- 損失函數小，與訓練資料的誤差較小

### 4.1 均方誤差(mean squared error): 越小越好
- $ E = \frac {1} {2} \sum_k (y_k-t_k)^2 $

```python
def mean_squared_error(y,t):
    return 0.5 * np.sum((y-t)**2)
```

### 4.2 交叉熵誤差(cross entropy error): 越小越好
- $ E = - \sum_k t_k \log y_k $

```python
def cross_entropy_error(y,t):
    delta = 1e-7 # 加上微小值 delta 避免 np.log(0) 會變成負無限大的 -inf
    return -np.sum(t*np.log(y+delta))
```

## 5 神經網路的梯度法(gradient method)
- 了解梯度法前，需先了解中央差分、偏微分及梯度
   - 中央差分: 如果用數值微分會有誤差，所以使用中央差分，也就是計算$(x+h)$與$(x-h)$的函數$f$差分，以減少誤差
   - 偏微分: 當算式有兩個以上變數，針對其中一個變數進行微分
   - 梯度: $(\frac{\partial f}{\partial x_0},\frac{\partial f}{\partial x_1})$ 這種把全部變數的偏微分當作向量來統一的方式

### 5.1 中央差分

```python
def numerical_diff(f,x):
    h = 1e-4 #0.0001
    return (f(x+h) - f(x-h)) / (2*h)
```

### 5.2 偏微分範例
以 $ f(x_0,x_1) = x^2_0 + x^2_1 $ 為例

```python
# 偏微分
def function_2(x):
    return x[0]**2 + x[1]**2
    # 或 return np.sum(x**2) 
```

### 5.3 梯度
- 梯度顯示的方向是在各個位置中，函數值減少最多的方向

```python
def numerical_gradient(f,x):  # f 為函數，x 為 NumPy 陣列
    h = 1e-4 #0.0001
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_val = x[idx]
        # 計算 f(x+h)
        x[idx] = tmp_val + h
        fxh1 = f(x)
        
        # 計算 f(x-h)
        x[idx] = tmp_val - h
        fxh2 = f(x)
        
        grad[idx] = (fxh1 - fxh2) / (2*h)
        x[idx] = tmp_val # 恢復原值
    return grad
```

```python
import numpy as np
numerical_gradient(function_2, np.array([3.0,4.0])) # 點 (3,4) 的梯度是 (6,8)
```

### 5.4 梯度法
- **用梯度找出函數的最小值的方法**。原理是反覆往梯度方向移動，逐漸減少函數值的方法，就是梯度法
- 梯度法是找出梯度為 0 的位置，但卻不見得是最小值，有可能是極小值或是鞍點
- 要注意的是，梯度是在各個地點，指出減少最多函數值的方向。無法保證梯度指出的位置使否真的是函數的最小值，以及這個方向是否正確，但朝向該方向，卻能減少最多函數值
- 當函數形成複雜的扭曲形狀時，可能進入平坦無法繼續學習的狀態，稱為「停滯期」
- 以算式表示:

    $ x_0 = x_0 - \eta \frac {\partial f} {\partial x_0} $

    $ x_1 = x_1 - \eta \frac {\partial f} {\partial x_1} $
    - $ \eta $ 代表更新的量，在神經網路的學習中，稱作**學習率(learning rate)**
    - 學習率決定了在一次學習中，要學習多少，更新多少參數
    - 學習率太大或太小，都無法達到「適當的位置」
- 找出最小值時，會稱為梯度下降法(gradient descent method)；找出最大值時，會稱為梯度上升法(gradient ascent method)

```python
# 梯度下降法
def gradient_descent(f, init_x, lr=0.01, step_num=100): 
    # f 是要最佳化的函數，init_x 是預設值，lr 是代表 learning rate 的學習率，step_num 是用梯度法重複的次數
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f, x)
        x -= lr * grad
    
    return x
```

#### 5.4.1 梯度法範例(續偏微分範例)
用梯度法求出 $ f(x_0,x_1) = x^2_0 + x^2_1 $ 的最小值

```python
init_x = np.array([-3.0,4.0]) # 假設預設值為(-3.0,4.0)
gradient_descent(function_2, init_x = init_x, lr = 0.1, step_num = 100)
# 結果為(-6.11110793e-10, 8.14814391e-10)，趨近於(0,0)，實際上最小值也是(0,0)
```

```python
#若學習率太大，會往大數值擴散 (設lr=10.0)
init_x = np.array([-3.0,4.0]) # 假設預設值為(-3.0,4.0)
gradient_descent(function_2, init_x = init_x, lr = 10.0, step_num = 100)
```

```python
#若學習率太小，幾乎不會更新就結束 (設lr=1e-10)
init_x = np.array([-3.0,4.0]) # 假設預設值為(-3.0,4.0)
gradient_descent(function_2, init_x = init_x, lr = 1e-10, step_num = 100)
```

### 5.5 神經網路的梯度範例
假設有個形狀只有2*3、權重為 $W$ 的神經網路，以 $L$ 代表損失函數。此時，可以用 $\frac {\partial L} {\partial W}$ 來表示梯度。算式如下

$ W = 
\begin{pmatrix}
w_{11} & w_{21} & w_{31}\\ 
w_{12} & w_{22} & w_{32}
\end{pmatrix} $

$ \frac {\partial L}{\partial W} = 
\begin{pmatrix}
\frac {\partial L}{\partial w_{11}} & \frac {\partial L}{\partial w_{21}} & \frac {\partial L}{\partial w_{31}}\\ 
\frac {\partial L}{\partial w_{12}} & \frac {\partial L}{\partial w_{22}} & \frac {\partial L}{\partial w_{32}}
\end{pmatrix} $

```python
import sys, os
import numpy as np
from functions import softmax, cross_entropy_error
from gradient import numerical_gradient


class simpleNet: # 把形狀為2*3的權重參數，當作一個實例變數
    def __init__(self):
        self.W = np.random.randn(2,3) # 以常態分佈初始化

    def predict(self, x):
        return np.dot(x, self.W) 

    def loss(self, x, t): # 計算損失函數用的方法，在引數 x 輸入資料，在 t 輸入正確答案標籤
        z = self.predict(x)
        y = softmax(z)
        loss = cross_entropy_error(y, t)

        return loss
```

```python
x = np.array([0.6, 0.9]) # 函數 f 的引數
t = np.array([0, 0, 1]) # 正確答案標籤

net = simpleNet()

f = lambda w: net.loss(x, t) 
dW = numerical_gradient(f, net.W) 
# numerical_gradient(f,x) 是計算梯度，net.W 是權重參數

print(dW)
# w11 約為 0.2，代表 w11 增加 h ，損失函數的值就會增加 0.2h
# w32 約為 -0.5，代表 w32 增加 h ，損失函數減少 0.5h
# 從減少損失函數的觀點來看， w32要往正值方向更新， w11 往負值方向更新即可，且 w32 的貢獻比 w11 大
```

## 6 執行學習演算法
- 神經網路的學習步驟:
   - 前提: 神經網路擁有適合的權重與偏權值，為了符合訓練資料，而調整權重與偏權值，這就稱作「學習」。神經網路是按照以下四個步驟來學習。
   - 步驟一: 小批次
      - 從續練資料中，隨機挑選出部分資料
      - 挑選出來的資料稱做小批次，這個步驟是以減少小批次的損失函數數值為目標
   - 步驟二: 計算梯度 (兩種方法: 數值微分、解析算式)
      - 為了減少小批次的損失函數，計算出各權重參數的梯度
      - 梯度會顯示出損失函數值減少最多的方向
   - 步驟三: 更新參數
      - 權重參數只往梯度方向進行微量更新
   - 步驟四: 重複
      - 重複執行步驟一、步驟二、步驟三
- 準確率梯度下降(stochastic gradient descent [SGD]): 針對隨機挑選出的資料進行梯度下降法
   - 準確率是指「準確隨機挑選」之意

### 6.1 範例: 以雙層神經網路為對象，使用MNIST資料集來學習
- 權重參數的預設值該如何設定，是決定神經網路學習成功與否的重要關鍵。這裡只簡單介紹權重是按照常態分佈的亂數來進行初始化，偏權值以 0 來初始化，後續會再詳細討論
- 為了要測試神經網路是不是能正確辨識訓練資料以外的資料，就要確認有沒有出現「過度學習」的狀況
   - 過度學習是指，只有訓練資料內的數字影像才能正確辨識，但是訓練資料以外的數字影像，就無法辨識

```python
# 引入 TwoLayerNet 類別
import sys, os
from functions import *
from gradient import numerical_gradient


class TwoLayerNet:

    def __init__(self, input_size, hidden_size, output_size, weight_init_std=0.01):
        # 權重初始化
        self.params = {}  # params 是維持神經網路參數的字典變數(實例變數)
        self.params['W1'] = weight_init_std * np.random.randn(input_size, hidden_size)
        self.params['b1'] = np.zeros(hidden_size)
        self.params['W2'] = weight_init_std * np.random.randn(hidden_size, output_size)
        self.params['b2'] = np.zeros(output_size)

    def predict(self, x): # 進行辨識 # x 是影像資料
        W1, W2 = self.params['W1'], self.params['W2']
        b1, b2 = self.params['b1'], self.params['b2']
    
        a1 = np.dot(x, W1) + b1
        z1 = sigmoid(a1)
        a2 = np.dot(z1, W2) + b2
        y = softmax(a2)
        
        return y
        
    # x:輸入資料, t:正確答案標籤
    def loss(self, x, t): # 計算損失函數
        y = self.predict(x)
        
        return cross_entropy_error(y, t)
    
    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)
        t = np.argmax(t, axis=1)
        
        accuracy = np.sum(y == t) / float(x.shape[0])
        return accuracy
        
    # 計算梯度: 用數值微分的方法
    # x:輸入資料, t:正確答案標籤
    def numerical_gradient(self, x, t): 
        loss_W = lambda W: self.loss(x, t)
        
        grads = {}
        grads['W1'] = numerical_gradient(loss_W, self.params['W1'])
        grads['b1'] = numerical_gradient(loss_W, self.params['b1'])
        grads['W2'] = numerical_gradient(loss_W, self.params['W2'])
        grads['b2'] = numerical_gradient(loss_W, self.params['b2'])
        
        return grads # 維持梯度的字典變數
```

```python
# 深入了解 TwoLayerNet
net = TwoLayerNet(input_size = 784, hidden_size = 100, output_size = 10)
net.params['W1'].shape # (784,100)
net.params['b1'].shape # (100,)
net.params['W2'].shape # (100,10)
net.params['b2'].shape # (10,)

import numpy as np
# np.random.rand(): 生成 [0, 1) 之間的隨機樣本
x = np.random.rand(100, 784) # 虛擬輸入資料 (100張)
t = np.random.rand(100, 10)  # 虛擬正確答案標籤 (100張)

grads = net.numerical_gradient(x, t) # 計算梯度

grads['W1'].shape # (784,100)
grads['b1'].shape # (100,)
grads['W2'].shape # (100,10)
grads['b2'].shape # (10,)
```

```python
# 使用 MNIST 資料集進行學習
# 隨著訓練資料的損失函數值減少，表示神經網路的學習順利進行
import sys, os
import numpy as np
import matplotlib.pyplot as plt
from mnist import load_mnist
# from two_layer_net import TwoLayerNet

# 引入 MNIST 函數
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, one_hot_label=True)

# 導入上方 TwoLayerNet 類別
network = TwoLayerNet(input_size=784, hidden_size=50, output_size=10)

# 超參數
iters_num = 10000  
train_size = x_train.shape[0]
batch_size = 100
learning_rate = 0.1

train_loss_list = []
train_acc_list = []
test_acc_list = []
#每 1 epoch 的重複次數
iter_per_epoch = max(train_size / batch_size, 1)  

for i in range(iters_num):
    # 取得小批次
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    # 計算梯度
    #grad = network.numerical_gradient(x_batch, t_batch)
    grad = network.gradient(x_batch, t_batch)
    
    # 記錄學習過程
    for key in ('W1', 'b1', 'W2', 'b2'):
        network.params[key] -= learning_rate * grad[key]
    
    loss = network.loss(x_batch, t_batch)
    train_loss_list.append(loss)
    
    #計算 1 epoch 的辨識準確度
    if i % iter_per_epoch == 0:   
        train_acc = network.accuracy(x_train, t_train)
        test_acc = network.accuracy(x_test, t_test)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)
        print("train acc, test acc | " + str(train_acc) + ", " + str(test_acc))

# 畫圖
markers = {'train': 'o', 'test': 's'}
x = np.arange(len(train_acc_list))
plt.plot(x, train_acc_list, label='train acc')
plt.plot(x, test_acc_list, label='test acc', linestyle='--')
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.ylim(0, 1.0)
plt.legend(loc='lower right')
plt.show() #從圖可知，兩種辨識的準確度沒有差別(幾乎重疊)，因此，學習沒有出現過度學習
```

## 7 誤差反向傳播法
- 在計算圖上由左往右進行計算的步驟，稱作正向傳播，其相反則是反向傳播(backward propagation)

### 7.1 加法與乘法節點
- 以購買蘋果與橘子的反向傳播範例說明加法與乘法節點的反向傳播


#### 7.1.1 加法層


```python
class AddLayer:
    def __init__(self):
        pass

    def forward(self, x, y):
        out = x + y

        return out

    def backward(self, dout):
        dx = dout * 1
        dy = dout * 1

        return dx, dy
```

#### 7.1.2 乘法層


```python
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None

    def forward(self, x, y):
        self.x = x
        self.y = y                
        out = x * y

        return out

    def backward(self, dout):
        dx = dout * self.y
        dy = dout * self.x

        return dx, dy
```

### 7.2 執行活化函數層
#### 7.2.1 ReLU 層: 如果正向傳播時的輸入值小於 0，反向傳播的值就會變成 0

```python
class Relu:
    def __init__(self):
        self.mask = None

    def forward(self, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0

        return out

    def backward(self, dout):
        dout[self.mask] = 0
        dx = dout

        return dx
```

#### 7.2.2 Sigmoid 層


```python
class Sigmoid:
    def __init__(self):
        self.out = None

    def forward(self, x):
        out = sigmoid(x)
        self.out = out
        return out

    def backward(self, dout):
        dx = dout * (1.0 - self.out) * self.out

        return dx
```

### 7.3 執行 Affine / Softmax 層
- Affine層: 就是利用神經網路的正向傳播，計算矩陣乘積，在幾何學中，稱作「仿射轉換(Affine Transformation)」
- 批次版 Affine 層


```python
class Affine:
    def __init__(self, W, b):
        self.W =W
        self.b = b
        
        self.x = None
        self.original_x_shape = None
        self.dW = None
        self.db = None

    def forward(self, x):
        self.original_x_shape = x.shape
        x = x.reshape(x.shape[0], -1)
        self.x = x

        out = np.dot(self.x, self.W) + self.b

        return out

    def backward(self, dout):
        dx = np.dot(dout, self.W.T)
        self.dW = np.dot(self.x.T, dout)
        self.db = np.sum(dout, axis=0)
        
        dx = dx.reshape(*self.original_x_shape)  
        return dx
```

### 7.4 Softmax-with-Loss 層
- 是指 Softmax 函數與損失函數交叉熵誤差(cross entropy error)
- 輸入影像是利用 Affine 層與 ReLU 層進行轉換，利用 Softmax 將 10 個輸入正規化

- Softmax-with-Loss 層的計算圖

- 上圖可以簡化如下，$(y_1-t_1,y_2-t_2,y_3-t_3)$ 是 Softmax 層的輸出與訓練資料的差分，表現出目前神經網路的輸出與訓練資料的誤差



```python
class SoftmaxWithLoss:
    def __init__(self):
        self.loss = None # 損失
        self.y = None # softmax 的輸出
        self.t = None # 訓練資料(one-hot vector)

    def forward(self, x, t):
        self.t = t
        self.y = softmax(x)
        self.loss = cross_entropy_error(self.y, self.t)
        
        return self.loss

    def backward(self, dout=1):
        batch_size = self.t.shape[0]
        if self.t.size == self.y.size: 
            dx = (self.y - self.t) / batch_size
        else:
            dx = self.y.copy()
            dx[np.arange(batch_size), self.t] -= 1
            dx = dx / batch_size
        
        return dx
```

### 7.5 執行對應誤差反向傳播法的神經網路
同6.1範例，但這邊使用層級傳播，就能達到處理便是結果(predict())、計算梯度(gradient())的目的

```python
import sys, os
import numpy as np
from layers import *
from common.gradient import numerical_gradient
from collections import OrderedDict


class TwoLayerNet:

    def __init__(self, input_size, hidden_size, output_size, weight_init_std = 0.01):
        # 權重初始化
        self.params = {}
        self.params['W1'] = weight_init_std * np.random.randn(input_size, hidden_size)
        self.params['b1'] = np.zeros(hidden_size)
        self.params['W2'] = weight_init_std * np.random.randn(hidden_size, output_size) 
        self.params['b2'] = np.zeros(output_size)

        # 產生各層
        self.layers = OrderedDict() # OrderDict 是有序字典。「有序」是指，可以記住新增至字典內的元素順序
        self.layers['Affine1'] = Affine(self.params['W1'], self.params['b1'])
        self.layers['Relu1'] = Relu()
        self.layers['Affine2'] = Affine(self.params['W2'], self.params['b2'])

        self.lastLayer = SoftmaxWithLoss()
    
    def predict(self, x):
        for layer in self.layers.values():
            x = layer.forward(x)
        
        return x
        
    #損失函數
    # x: 輸入資料, t: 訓練資料(正確資料標籤)
    def loss(self, x, t):
        y = self.predict(x)
        return self.lastLayer.forward(y, t)
    
    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)
        if t.ndim != 1 : t = np.argmax(t, axis=1)
        
        accuracy = np.sum(y == t) / float(x.shape[0])
        return accuracy
        
    # x: 輸入資料, t: 訓練資料(正確資料標籤)
    def numerical_gradient(self, x, t):
        loss_W = lambda W: self.loss(x, t)
        
        grads = {}
        grads['W1'] = numerical_gradient(loss_W, self.params['W1'])
        grads['b1'] = numerical_gradient(loss_W, self.params['b1'])
        grads['W2'] = numerical_gradient(loss_W, self.params['W2'])
        grads['b2'] = numerical_gradient(loss_W, self.params['b2'])
        
        return grads
        
    def gradient(self, x, t):
        # forward
        self.loss(x, t)

        # backward
        dout = 1
        dout = self.lastLayer.backward(dout)
        
        layers = list(self.layers.values())
        layers.reverse()
        for layer in layers:
            dout = layer.backward(dout)

        # 設定
        grads = {}
        grads['W1'], grads['b1'] = self.layers['Affine1'].dW, self.layers['Affine1'].db
        grads['W2'], grads['b2'] = self.layers['Affine2'].dW, self.layers['Affine2'].db

        return grads
```

- 確認數值微分計算結果與誤差反向傳播法計算出來的梯度一致(或近似)，這就稱為梯度檢查(gradient check)

```python
import sys, os
import numpy as np
from mnist import load_mnist
from two_layer_net import TwoLayerNet

# 載入資料
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, one_hot_label=True)

network = TwoLayerNet(input_size=784, hidden_size=50, output_size=10)

x_batch = x_train[:3]
t_batch = t_train[:3]

grad_numerical = network.numerical_gradient(x_batch, t_batch)
grad_backprop = network.gradient(x_batch, t_batch)

# 計算各權重的絕對誤差平均值
for key in grad_numerical.keys():
    diff = np.average( np.abs(grad_backprop[key] - grad_numerical[key]) )
    print(key + ":" + str(diff))
    
# 從結果可知，數值微分與誤差反向傳播法計算出來的梯度誤差非常小
```

## 8 與學習有關的技巧

### 8.1 更新參數
- 神經網路的學習目的，就是找出可以盡量縮小損失函數的參數，亦即找出最佳參數，解決這種問題的做法，稱為最佳化(optimazation)
- 以下要介紹四種手法: SGD、Momentum、AdaGrad、Adam

#### 8.1.1 SGD
- 隨機梯度下降法(Stochastic Gradient Descent)
- 前面已經介紹過 SGD，雖然他很單純、易於執行，但若是函數形狀沒有等向性、非延伸形狀的函數，就會以沒有效率的路徑來進行探索
- $W \leftarrow W-\eta \frac{\partial L}{\partial W}$

```python
class SGD:

    def __init__(self, lr=0.01):
        self.lr = lr
        
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key] 
```

#### 8.1.2 Momentum
- 是依照球在網內來回滾動的物理定律為基準來移動
- $v \leftarrow \alpha v - \eta \frac{\partial L}{\partial W}$

  $W \leftarrow W + v$
  
   - $\alpha v$ 是當物體沒有受力時，逐漸減速的功能 (通常 $\alpha$ 設為0.9) 

```python
class Momentum:

    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum # alpha
        self.v = None
        
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():                                
                self.v[key] = np.zeros_like(val)
                
        for key in params.keys():
            self.v[key] = self.momentum*self.v[key] - self.lr*grads[key] 
            params[key] += self.v[key]
```

#### 8.1.3 AdaGrad
- 是依照各個參數，準備客製值，調整適應的更新步驟
- 學習率衰減(learning rate decay)
   - 隨著學習過程來縮小學習率的方法
   - 學習率太小，會花費太多時間在學習上；學習率太大，會往外擴散，無法正確學習
- $h \leftarrow h + \frac{\partial L}{\partial W} \odot \frac{\partial L}{\partial W}$

  $W \leftarrow W - \eta \frac {1} {\sqrt{h}} \frac{\partial L}{\partial W}$
  
   - $\odot$ 是矩陣各元素相乘
   - 更新參數時，乘上$\frac {1} {\sqrt{h}}$，調整學習規模，也就是經常變動(放大)的元素，學習率變小的意思
   - 由於經常變動的參數學習率會逐漸變小，所以我們可以針對參數的各個元素，執行學習率衰減

```python
class AdaGrad:

    def __init__(self, lr=0.01):
        self.lr = lr
        self.h = None
        
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
            
        for key in params.keys():
            self.h[key] += grads[key] * grads[key]
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7) 
            # 加上 1e-7 是為了避免當 self.h[key] 出現0時，發生除以0的問題
```

#### 8.1.4 Adam
- 其雛形就是融合了 Momentum 與 AdaGrad 兩種手法的優點，另外，還能進行超參數的「偏權值校正(偏離校正)」
- Adam 的三個超參數，一個是學習率，另外兩個是第一時刻係數 $\beta_1$ 以及第二時刻係數 $\beta_2$。標準設定值是 $\beta_1 = 0.9$、$\beta_2 = 0.999$，此設定值可以解決大多數的情況

```python
class Adam:

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads):
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        lr_t  = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)         
        
        for key in params.keys():
            #self.m[key] = self.beta1*self.m[key] + (1-self.beta1)*grads[key]
            #self.v[key] = self.beta2*self.v[key] + (1-self.beta2)*(grads[key]**2)
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])
            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)
            
            #unbias_m += (1 - self.beta1) * (grads[key] - self.m[key]) # correct bias
            #unbisa_b += (1 - self.beta2) * (grads[key]*grads[key] - self.v[key]) # correct bias
            #params[key] += self.lr * unbias_m / (np.sqrt(unbisa_b) + 1e-7)
```

#### 8.1.5 如何決定要用何種更新手法?
- 每種手法都有其擅長與不擅長解決的問題，目前 SGD 與 Adam 最常被拿來使用
- 一般而言，與 SGD 相比，其他三種手法的學習速度的確較快，有時最後的辨識效能也比較好

#### 8.1.6 使用 MNIST資料集比較更新手法
從結果可知，AdaGrad的學習速度稍微快一些

### 8.2 權重的預設值
- 權重衰減(Weight decay)是以縮小權重參數值為目的，進行學習的手法。縮小權重值，可以避免過度學習。
    - 如果想要縮小權重值，預設值也盡量從小數值開始，稱為正攻法
- 權重不可為零，也不可為均一值
    - 為了避免「權重變成均一值」，正確來說是要破壞權重的對稱結果，需要隨機的預設值
- 權重的預設值會影響神經網路的學習成功與否，卻是常常被忽略的關鍵

#### 8.2.1 隱藏層的活性化分布
- 活性化(Activation): 活化函數之後的輸出資料
- 梯度消失(gradient vanishing): 在偏向 0 與 1 的資料分布中，反向傳播的梯度值會逐漸變小、消失
- 各層的活性化分布需要擁有適當的廣泛度。因為自各層傳遞具有適當多樣性的資料，才能有效進行神經網路的學習；相對來說，單一偏差的資料會造成梯度消失、「表現力受限」等問題，因而無法順利學習
- 權重預設值
    - 活化函數為 sigmoid 或 tanh 等 S 型曲線，使用「Xavier預設值」
    - 活化函數為 ReLU 時，使用「He預設值」

#### 8.2.2 Xavier預設值
- 假設上層節點有$n$個，要使用具有 $\frac {1} {\sqrt{n}$ 標準差的常態分佈來初始化
- sigmoid 或 tanh 函數的差別:
    - 如果上層的分布形狀並非常態分布，而是有點不規則，則使用 tahn 函數(雙曲線函數)取代 sigmoid 函數來改善
    - tahn 函數在原點(0,0)呈現對稱 S 型曲線，而 sigmoid 函數則是在(0,0.5)呈現對稱

#### 8.2.3 He預設值
- 使用 ReLU 函數時，最好利用其專用的預設值，也就是「He預設值」
- 假設上層節點有$n$個，使用標準差為 $\sqrt {\frac {2} {n}}$ 的常態分佈來初始化

#### 8.2.4 利用 MNIST 資料集比較權重預設值
- 比較「std=0.01」、「Xavier預設值」、「He預設值」等三種情況
- 當「std=0.01」時，完全不會學習，因為正向傳播傳遞的數值非常小(集中在0的資料)，使得反向傳播時，計算出來的梯度也會變小，因而幾乎不會更新權重。相對來說，使用 Xavier 預設值或 He 預設值時可以順利學習，其中又以 He 預設值的學習速度比較快

```python
import numpy as np
import matplotlib.pyplot as plt
from mnist import load_mnist
from util import smooth_curve
from multi_layer_net import MultiLayerNet
from optimizer import SGD

# 0: 讀入 MNIST 資料
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

train_size = x_train.shape[0]
batch_size = 128
max_iterations = 2000


# 1: 實驗設定
weight_init_types = {'std=0.01': 0.01, 'Xavier': 'sigmoid', 'He': 'relu'} # 不同種的權重預設值
optimizer = SGD(lr=0.01) # 用 SGD 方法更新參數

networks = {}
train_loss = {}
for key, weight_type in weight_init_types.items():
    networks[key] = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100],
                                  output_size=10, weight_init_std=weight_type)
    train_loss[key] = []


# 2: 開始訓練資料
for i in range(max_iterations):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    for key in weight_init_types.keys():
        grads = networks[key].gradient(x_batch, t_batch)
        optimizer.update(networks[key].params, grads)
    
        loss = networks[key].loss(x_batch, t_batch)
        train_loss[key].append(loss)
    
    if i % 100 == 0:
        print("===========" + "iteration:" + str(i) + "===========")
        for key in weight_init_types.keys():
            loss = networks[key].loss(x_batch, t_batch)
            print(key + ":" + str(loss))


# 3.グラフの描画==========
markers = {'std=0.01': 'o', 'Xavier': 's', 'He': 'D'}
x = np.arange(max_iterations)
for key in weight_init_types.keys():
    plt.plot(x, smooth_curve(train_loss[key]), marker=markers[key], markevery=100, label=key)
plt.xlabel("iterations")
plt.ylabel("loss")
plt.ylim(0, 2.5)
plt.legend()
```

### 8.3 Batch Normalization
- 用以調整各層的活性化分布，變得具有適當廣泛性。在活化函數之前(或之後)插入這種處理步驟，可以減少資料分布的偏差。

- 其運行過程，是在進行學習時，以小批次為單位，依照各個小批次來進行正規化(標準化)，也就是讓資料分配在平均為 0 標準差為 1 的分布下
   - 平均: $ \mu_B \leftarrow \frac {1}{m}\sum_{i=1}^{m}x_i $
   - 標準差: $ \sigma_B^2 \leftarrow \frac {1}{m}\sum_{i=1}^{m}(x_i-\mu_B)^2 $
   - 轉換: $ \hat{x_i} \leftarrow \frac{x_i-\mu_B}{\sqrt{\sigma_B^2-\varepsilon}} $
      - $\varepsilon$ 是個微小值，防止除以 0
   - 標準化後的縮放平移: $ y_i \leftarrow \gamma \hat{x_i} + \beta $
      - $\gamma$與$\beta$是參數，最初從$\gamma=1$、$\beta=0$開，藉由學習調整成適當的值 
- 優點:
   - 可以快速學習(因為能增加學習率)
   - 不會過度依賴預設值(不會對預設值產生過度反應)
   - 控制過度學習(減少 Dropout 等必要性)

### 8.4正規化
- 正規化項目: 假設權重$W=(w_1,w_2,...,w_n)$
   - L1 norm: $\sqrt{w_1^2+w_2^2+...+w_n^2}$
   - L2 norm: $\left | w_1 \right |+\left | w_2 \right |+...+\left | w_n \right |$
   - L $\infty $ norm: 各元素中，最大的絕對值

#### 8.4.1 過度學習
- 過度學習(overfitting): 過度適應訓練資料，而無法順利對應不含訓練資料的其他資料。
- 原因:
   1. 擁有大量參數，表現力高的模型
   2. 訓練資料太少
- 控制過度學習的方法
   - Weight decay
   - Dropout

#### 8.4.2 Weight decay (權重衰減)
- 在學習的過程中，針對擁有較大的權重部分，課以罰金，藉此控制過度學習。
- 針對全部的權重，將$\frac{1}{2}\lambda W^2$加入損失函數中。因此，計算權重的梯度時，將誤差反向傳播法的結果與正規化項目的微分$\lambda W$相加

#### 8.4.3 Dropout
- 當神經網路變複雜時，單憑 Weight decay，很難解決過度學習的問題
- Dropout是一邊消除神經元，一邊學習的手法


```python
'''
每次正向傳播時，把刪除的神經元當作 False 儲存在 self.mask 中。
self.mask 是隨機產生和 x 相同形狀的陣列，只有這個值比 dropout_ratio 大的元素當作 True。
'''
class Dropout:
    """
    http://arxiv.org/abs/1207.0580
    """
    def __init__(self, dropout_ratio=0.5):
        self.dropout_ratio = dropout_ratio
        self.mask = None

    def forward(self, x, train_flg=True):
        if train_flg:
            self.mask = np.random.rand(*x.shape) > self.dropout_ratio
            return x * self.mask
        else:
            return x * (1.0 - self.dropout_ratio)

    def backward(self, dout):
        return dout * self.mask
```

### 8.5 驗證超參數
- 超參數(hyper-parameter)是指，各層的神經元數量、批次大小、參數更新時的學習率、Weight decay
- 如果沒有妥善設定超參數，就會變成效能不佳的模型

#### 8.5.1 驗證資料
- 不可使用測試資料來評估超參數的效能，若使用測試資料調整超參數，超參數的值會對測試資料產生過度學習
- 調整超參數時，必須準備超參數專用的確認資料，稱為驗證資料(validation data)
- 取驗證資料最簡單的方式就是從訓練資料中，先分出 20% 的資料，當作驗證資料

#### 8.5.2 超參數最佳化
- 關鍵在於，逐漸縮小「優良」超參數存在的範圍。
- 步驟: 逐漸縮小範圍是指，剛開始設定概略的範圍，從該範圍中，隨機選出超參數(取樣)，利用樣本值評估辨識準確度。反覆重複數次，觀察辨識準確度的結果，從結果中，縮小「優良」超參數的範圍。重複執行以上步驟，即可逐漸篩選出適當的超參數範圍。縮小至一定程度後，從該範圍中，選出一個超參數的值。
   - 概略的範圍: 以 10 的次方來設定範圍 e.g. 0.001 到 1000
- 超參數最佳化需要花費許多時間，因此在尋找超參數時，必須盡早排除可能不適合的超參數

## 9 卷積神經網路 (convolutional neural network: CNN)
- 使用於影像辨識、聲音辨識

### 9.1 整體結構
- 跟前面的神經網路類似，只是多了「卷積層(Convolution 層)」與「池化層(Pooling 層)」

- 在輸出層使用的是「Affine - Softmax」組合

### 9.2 卷積層

#### 9.2.1 全連接層的問題
- 前面介紹的全連接神經網路，使用了全連接層(Affine 層)。全連接層連接相鄰各層的所有神經元，可以決定任意輸出數量。
- 全連接層的問題就是會忽略資料的形狀。例如，輸入資料為影像，影像通常是含有水平、垂直、色板方向的三維形狀，此三維形狀包含了重要的空間資料，若用前面介紹的全連接神經網路就無法保留此特性，但如果使用卷積層，就能維持形狀。
- 如果是影像，輸入資料可以當作三維資料來處理，對下一層輸出同樣是三維的資料。因此，CNN可以正確了解影像等含有形狀的資料
- CNN的卷積層輸出入資料稱作特徵圖(feature map)。可再細分為輸入特徵圖(input feature map)及輸出特徵圖(output feature map)

#### 9.2.2 卷積運算

- 在各個位置呈上濾鏡的元素與對應輸入的元素，並計算總合(這個部分也稱作積和運算)。再將結果儲存在對應輸出的位置
- 全連接的神經網路中，除了權重參數之外，還有偏權值。CNN的濾鏡參數對應的是前面提到的權重，而且在CNN之中，也有偏權值

#### 9.2.3 填補(padding)

- 進行卷積處理之前，必須在輸入資料的周圍填上固定的資料(例如0)，這個動作稱作填補
- 填補的值可以為任意整數
- 使用填補是為了調整輸出大小。如果每次進行積卷運算時，空間就會縮小的化，到了某個階段，輸出大小就會變成 1，而無法再進行卷積運算，使用填補就是為了避免這種情況

#### 9.2.4 步幅(stride)

這裡假設輸入大小(H,W)，濾鏡大小為(FH,FW)，輸出大小為(OH,OW)，填補為P，步幅為S

$OH=\frac{H+2P-FH}{S}+1$

$OW=\frac{W+2P-FW}{S}+1$

一定要讓以上兩個算式整除。假如輸出大小無法整除，就必須採取輸出錯誤對策(例如變成最接近的整數)

#### 9.2.5 三維資料的卷積運算
如果色板方向有多個特徵圖時，要依照各個色板進行輸入資料與濾鏡的卷積運算，把結果相加，獲得一個輸出


#### 9.2.6 用區塊來思考
- 把三維資料當作多維陣列時，會依照(channel,height,width)的順序來顯示

- 如上圖所示，套用FN個濾鏡，也會產生FN個輸出的特徵圖

#### 9.2.7 批次處理
- 如同前面所說的神經網路，卷積運算也希望能支援批次處理，因此把流動在各層的資料儲存成四維資料，並依照(batch_num,channel,height,width)的順序來儲存資料
- 雖然在網路中傳遞四維資料，但這代表針對N個資料來進行卷積運算


### 9.3 池化層
- 池化層是縮小垂直、水平空間的運算
- 一般而言，持化的視窗大小與步幅設定成相同數值
- 池化層有兩種，大多使用最大池化:
   - 最大池化(max pooling): 從目標區域中，取得最大值
   - 平均池化(average pooling): 計算目標區域的平均值
- 特色
   - 沒有學習參數: 池化只進行從目標區域取得最大值(或平均值)的處理，所以沒有必須學習的參數存在
   - 色板數量不變
   - 對微小位置變化很穩健(robust)

### 9.4 執行卷積層 / 池化層

#### 9.4.1 im2col
- im2col 函數是可以針對套用濾鏡(權重)，輕鬆展開輸入資料的函數
- im2col = image to column，也就是將影像轉換成陣列
- 進行卷積運算時，幾乎濾鏡區域都會重疊。套用濾鏡的區域重疊時，利用im2col展開，展開後的元素數量，會變得比原本區塊的元素數量還多
- 缺點是會占用比較多記憶體


```python
def im2col(input_data, filter_h, filter_w, stride=1, pad=0):
    """

    Parameters
    ----------
    input_data : 由(資料數量, 色板, 高度, 寬度)的陣列形成的輸入資料
    filter_h : 濾鏡的高度
    filter_w : 濾鏡的寬度
    stride : 步幅
    pad : 填補

    Returns
    -------
    col : 2維陣列
    """
    N, C, H, W = input_data.shape
    out_h = (H + 2*pad - filter_h)//stride + 1
    out_w = (W + 2*pad - filter_w)//stride + 1

    img = np.pad(input_data, [(0,0), (0,0), (pad, pad), (pad, pad)], 'constant')
    col = np.zeros((N, C, filter_h, filter_w, out_h, out_w))

    for y in range(filter_h):
        y_max = y + stride*out_h
        for x in range(filter_w):
            x_max = x + stride*out_w
            col[:, :, y, x, :, :] = img[:, :, y:y_max:stride, x:x_max:stride]

    col = col.transpose(0, 4, 5, 1, 2, 3).reshape(N*out_h*out_w, -1)
    return col
```

#### 9.4.2 執行卷積層
- 卷積層的初始化方法是把濾鏡(權重)、偏權值、步幅、填補當作引數來取得
- reshape時，設定成-1，可以整合元素數量，是多維陣列的元素數量完美配合。例如，(10,3,5,5)形狀的陣列，全部的元素數量共有 750 個，假設(10,-1)，就能調整成形狀為(10,75)的陣列。
- transpose 是更換多維陣列各軸順序的函數


```python
class Convolution:
    def __init__(self,W,b,stride=1,pad=0):
        self.W = W
        self.b = b
        self.stride = stride
        self.pad = pad
    '''
    FN = Fliter Number(濾鏡的數量)
    C = Channel
    FH = Fliter Height
    FW = Filter Width
    '''
    def forward(self,x):
        FN,C,FH,FW = self.W.shape 
        N,C,H,W = x.shape
        out_h = int(1 + (H + 2*self.pad - FH) / self.stride)
        out_w = int(1 + (W + 2*self.pad - FW) / self.stride)

        # 卷积层实现的重要部分
        col = im2col(x,FH,FW,self.stride,self.pad)
        col_W = self.W.reshape(FN,-1).T # 滤波器的展开
        out =  np.dot(col,col_W) + self.b
        
        out = out.reshape(N,out_h,out_w,-1).transpose(0,3,1,2)
        
        return out
```

#### 9.4.3 執行池化層
- 執行過程
   1. 展開輸入資料
   2. 計算每列最大值
   3. 調整成適當的輸出大小


```python
class Pooling:
    def __init__(self,pool_h,pool_w,stride=1,pad=0):
        self.pool_h = pool_h
        self.pool_w = pool_w
        self.stride = stride
        self.pad = pad
    def forward(self, x):
        N, C, H, W = x.shape
        out_h = int(1 + (H - self.pool_h) / self.stride)
        out_w = int(1 + (W - self.pool_w) / self.stride)
        
        # 展開(1)
        col = im2col(x, self.pool_h, self.pool_w, self.stride, self.pad)
        col = col.reshape(-1, self.pool_h*self.pool_w)
        
        # 最大值(2)
        out = np.max(col, axis=1)
        
        # 調整形狀(3)
        out = out.reshape(N, out_h, out_w, C).transpose(0, 3, 1, 2)
        
        return out
```

### 9.5 執行 CNN
網路結構如下


```python
# coding: utf-8
import pickle
import numpy as np
from collections import OrderedDict
from layers import *
from gradient import numerical_gradient


class SimpleConvNet:
    """SimpleConvNet的結構

    conv - relu - pool - affine - relu - affine - softmax
    
    Parameters
    ----------
    input_size : 輸入大小（MNIST的情况下為784）
    hidden_size_list : 隱藏層的神經元數量的列表（e.g. [100, 100, 100]）
    output_size : 輸出大小（MNIST的情况下為10）
    activation : 'relu' or 'sigmoid'
    weight_init_std : 指定權重的標準差（e.g. 0.01）
        指定'relu'或'he'的情况下設定“He的初始值”
        指定'sigmoid'或'xavier'的情况下設定“Xavier的初始值”
    """
    # input_dim―输入数据的维度：（ 通道，高，长）
    def __init__(self, input_dim=(1, 28, 28), 
                 conv_param={'filter_num':30, 'filter_size':5, 'pad':0, 'stride':1}, 
                 hidden_size=100, output_size=10, weight_init_std=0.01):
        # 取出卷積層的超參數方便使用
        filter_num = conv_param['filter_num']
        filter_size = conv_param['filter_size']
        filter_pad = conv_param['pad']
        filter_stride = conv_param['stride']
        input_size = input_dim[1]

        # 計算卷積層的輸出大小
        conv_output_size = (input_size - filter_size + 2*filter_pad) / filter_stride + 1
        pool_output_size = int(filter_num * (conv_output_size/2) * (conv_output_size/2))
        
        # 初始化權重參數
        # 學習所需的參數是第一層的卷積層和剩餘兩個全連接層的權重和偏權值
        self.params = {}
        self.params['W1'] = weight_init_std * \
                            np.random.randn(filter_num, input_dim[0], filter_size, filter_size)
        self.params['b1'] = np.zeros(filter_num)
        self.params['W2'] = weight_init_std * \
                            np.random.randn(pool_output_size, hidden_size)
        self.params['b2'] = np.zeros(hidden_size)
        self.params['W3'] = weight_init_std * \
                            np.random.randn(hidden_size, output_size)
        self.params['b3'] = np.zeros(output_size)
        
        # 生成層
        # 產生必要的層級
        # 從最前面開始按順序向有序字典（OrderedDict）的 layers中，增加層數
        self.layers = OrderedDict()
        self.layers['Conv1'] = Convolution(self.params['W1'], self.params['b1'],
                                           conv_param['stride'], conv_param['pad'])
        self.layers['Relu1'] = Relu()
        self.layers['Pool1'] = Pooling(pool_h=2, pool_w=2, stride=2)
        self.layers['Affine1'] = Affine(self.params['W2'], self.params['b2'])
        self.layers['Relu2'] = Relu()
        self.layers['Affine2'] = Affine(self.params['W3'], self.params['b3'])

        # 只有最后的 SoftmaxWithLoss 層加上另一個變數 lastLayer 
        self.last_layer = SoftmaxWithLoss()
        
    # 進行推理的 predict 方法
    def predict(self, x):
        for layer in self.layers.values():
            x = layer.forward(x)

        return x
    # 計算損失函數 loss 方法
    def loss(self, x, t):  # x 是輸入資料，t 是訓練標籤
        """求损失函数
        参数x是输入数据、t是教师标签
        """
        y = self.predict(x)
        return self.last_layer.forward(y, t)
    
    def accuracy(self, x, t, batch_size=100):
        if t.ndim != 1 : t = np.argmax(t, axis=1)
        
        acc = 0.0
        
        for i in range(int(x.shape[0] / batch_size)):
            tx = x[i*batch_size:(i+1)*batch_size]
            tt = t[i*batch_size:(i+1)*batch_size]
            y = self.predict(tx)
            y = np.argmax(y, axis=1)
            acc += np.sum(y == tt) 
        
        return acc / x.shape[0]
    
    def numerical_gradient(self, x, t):
        """求梯度（誤差反向傳播法）

        Parameters
        ----------
        x : 輸入資料
        t : 訓練標籤

        Returns
        -------
        具有各層的梯度的字典變量
            grads['W1']、grads['W2']、...是各層的權重
            grads['b1']、grads['b2']、...是各層的偏權值
        """
        loss_w = lambda w: self.loss(x, t)

        grads = {}
        for idx in (1, 2, 3):
            grads['W' + str(idx)] = numerical_gradient(loss_w, self.params['W' + str(idx)])
            grads['b' + str(idx)] = numerical_gradient(loss_w, self.params['b' + str(idx)])

        return grads
 
    # 利用誤差反向傳播法，計算梯度
    def gradient(self, x, t):
        """求梯度（誤差反向傳播法）

        Parameters
        ----------
        x : 輸入資料
        t : 訓練標籤

        Returns
        -------
        具有各層的梯度的字典變量
            grads['W1']、grads['W2']、...是各層的權重
            grads['b1']、grads['b2']、...是各層的偏權值
        """
        # forward
        self.loss(x, t)

        # backward
        dout = 1
        dout = self.last_layer.backward(dout)

        layers = list(self.layers.values())
        layers.reverse()
        for layer in layers:
            dout = layer.backward(dout)

        # 設定
        # 在 grads 字典中，儲存各權重參數的梯度
        grads = {}
        grads['W1'], grads['b1'] = self.layers['Conv1'].dW, self.layers['Conv1'].db
        grads['W2'], grads['b2'] = self.layers['Affine1'].dW, self.layers['Affine1'].db
        grads['W3'], grads['b3'] = self.layers['Affine2'].dW, self.layers['Affine2'].db

        return grads

    def save_params(self, file_name="params.pkl"):
        params = {}
        for key, val in self.params.items():
            params[key] = val
        with open(file_name, 'wb') as f:
            pickle.dump(params, f)

    def load_params(self, file_name="params.pkl"):
        with open(file_name, 'rb') as f:
            params = pickle.load(f)
        for key, val in params.items():
            self.params[key] = val

        for i, key in enumerate(['Conv1', 'Affine1', 'Affine2']):
            self.layers[key].W = self.params['W' + str(i+1)]
            self.layers[key].b = self.params['b' + str(i+1)]
```

#### 9.6.1 視覺化第一層權重
- 由學習前的濾鏡進行了隨機初始化，所以黑白深淺沒有規則性。然而，結束學習的濾鏡變成了有規則性的影像。
- 藉由學習，更新成具有規則性的濾鏡，包括伴隨由白到黑漸層變化的濾鏡、具有塊狀區域(這裡稱作「塊(blob)」)的濾鏡等。
- 卷積層的濾鏡可以擷取出邊界與塊狀等原始資料，而就是用前面說明的CNN把這種原始資料傳遞給下層。

```python
import numpy as np
import matplotlib.pyplot as plt
from simple_convnet import SimpleConvNet

def filter_show(filters, nx=8, margin=3, scale=10):
    """
    c.f. https://gist.github.com/aidiary/07d530d5e08011832b12#file-draw_weight-py
    """
    FN, C, FH, FW = filters.shape
    ny = int(np.ceil(FN / nx))

    fig = plt.figure()
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

    for i in range(FN):
        ax = fig.add_subplot(ny, nx, i+1, xticks=[], yticks=[])
        ax.imshow(filters[i, 0], cmap=plt.cm.gray_r, interpolation='nearest')
    plt.show()


network = SimpleConvNet()
# 隨機進行初始化後的權重
filter_show(network.params['W1'])

# 學習後的權重
network.load_params("params.pkl")
filter_show(network.params['W1'])
```
